```python
# Load an image from file
image = cv2.imread('/Users/dianaterraza/Desktop/workspace_python/Cat_November_2010-1a.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Draw a rectangle on the image (top-left corner at (50, 50),
# bottom-right corner at (200, 200))
cv2.rectangle(image, (50, 50), (200, 200), (0, 255, 0), thickness=2)

# Display the image with a rectangle
plt.imshow(image)
```

```python
import requests # importing the library we need to use an API

# the url below is a link to the document. We'll send to to the API
url = 'https://pixeljets.com/blog/content/images/2021/12/scan_your_pdf.png'

# api below is a link that we define to get results from the OCR.
# To complete the link we need to pass in
# the personal API key and the url to the image

key = 'K85577689588957'

api = f'https://api.ocr.space/parse/imageurl?apikey={key}&url={url}'

# we use the requests library to send a request to an API via the link above
ocr_results = requests.get(api)

# we extract the results as a JSON file
extracted_info = ocr_results.json()

# we extract the text from the result and print it
print(extracted_info['ParsedResults'][0]['ParsedText'].replace("\r\n", ""))
```
