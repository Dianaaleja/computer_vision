Apply gamma correction to our image 

```python
img = cv2.imread('/Users/dianaterraza/Desktop/workspace_python/476C9B9C-19EE-42D5-A5E5-0DCFCDEA29AF.png')
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)/255.0 #max value for every channel
plt.imshow(img)
```

```python
def gamma_correction(img, gamma):
  rows, cols, channels = img.shape
  out = np.zeros_like(img)
  for r in range(rows):
    for c in range(cols):
      for ch in range(channels):
        out[r,c,ch] = img[r,c,ch]**gamma
  return out
```

```python
plt.subplot(131), plt.imshow(img), plt.title('Original')
plt.subplot(132), plt.imshow(gamma_correction(img, 1.5)), plt.title('Gamma = 1.5')
plt.subplot(133), plt.imshow(gamma_correction(img, 1/3)), plt.title('Gamma = 1/3')
```

* Pixels : rows and columns 
* 0-255 from white to black
* RGB order the colours 
* apply a func that change the number of the channels.
